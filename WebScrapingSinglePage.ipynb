```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://www.popvortex.com/music/charts/top-100-songs.php'
```

```python
response = requests.get(url)
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
soup
```

```python
body > div.container > div:nth-child(4) > div.col-xs-12.col-md-8 > div.chart-wrapper
# song name 
#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > cite

# artist
#chart-position-1 > div.chart-content.col-xs-12.col-sm-8 > p > em
```

```python
songs_selection = soup.select(".title")
artists_selection = soup.select(".artist")
```

```python
songs = [tag.text for tag in songs_selection]
artists = [tag.text for tag in artists_selection]
```

```python
df = pd.DataFrame({'Song': songs, 'Artist': artists})
```

```python
df.to_csv('top100.csv')
```
